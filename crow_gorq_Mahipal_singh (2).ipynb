```python

llm = ChatGroq(model="gemma-7b-it", groq_api_key=os.environ.get('GROQ_API_KEY'))
# Use os.environ.get() for more robust key retrieval

print(llm.invoke("hi"))
search_tool = SerperDevTool()
```

```python
grader = Agent(
    role="Grader",
    goal="Accurately and efficiently evaluate and grade student submissions",
    backstory=(
        "As an expert educator and evaluator, your mission is to assess"
        "student work with fairness and precision. You leverage advanced"
        "language models to ensure each student's effort is recognized"
        "and accurately graded, providing constructive feedback to foster"
        "learning and improvement."
    ),
    llm=llm,
    max_iter=5,
    tools=[search_tool],
    allow_delegation=False,
    verbose=True
)
```

```python
grading_task = Task(
    description=(
        "Evaluate and grade student submissions based on their content,"
        " clarity, and adherence to the assignment guidelines. Provide"
        " constructive feedback for each submission to help students"
        " improve their work."
    ),
    expected_output='A detailed grade and feedback report for each student submission.',
    tools=[search_tool],
    agent=grader,
    async_execution=False,
    output_file="grades_and_feedback.json"
)
```

```python
crew = Crew(
    agents = [grader ],
    tasks = [grading_task],
    process = Process.sequential,
    max_rpm = 3,
    cache=True
)
```

```python
result = crew.kickoff(inputs={'subject_area':"Genrative AI"})
print(result)
```
